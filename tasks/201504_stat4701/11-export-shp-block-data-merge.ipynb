```python
df = pd.read_csv(pj+'output/all/nycb2010_taxi_2013_stats.csv', dtype={'geoid':object}).rename(columns=lambda x: x.lower())

df = df[['geoid','avgbrdist','medbrdist','stdbrdist','count','areasqft']]

bl = pd.read_csv(pj+'processing/buildings/cb2010_building_0913_int_describe_cnt.csv', dtype={'geoid':object})

ba = pd.read_csv(pj+'processing/buildings/cb2010_building_0913_int_dis.csv', dtype={'geoid':object})
ba = ba[['geoid','bldgareaft','pctcbbldg']]

dfb = pd.read_csv(pj+'input/building/bldg_dist_height.csv').rename(columns=lambda x: x.lower())
dfb = dfb[['geoid','building_block_int_dis_tbl_bulkdens']]
dfb = dfb.fillna(0)  
dfb['geoid'] = dfb['geoid'].astype(str)
dfb.columns = ['geoid','dist_bldg_hght']


df = df.merge(bl, how='left', on='geoid').merge(ba, how='left', on='geoid').merge(dfb, how='left', on='geoid')

df.to_csv(pj+'output/all/nycb2010_taxi_2013_stats_bldg_cnt_pctcbbldg.csv', index=False)

df
```
